```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
os.chdir ('C:\\Noble\\Training\\Top Mentor\\Training\\Presentation\\Project\\Project -3 Wheat Export')
df = pd.read_csv('Wheat Export.csv',index_col = 0, thousands  = ',')
df
```

```python
df1= df
df1.index.names = ['stations']
df1.columns.names = ['Years']
df1
```

```python
df.isna().sum()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df1)
```

```python
PCA1 = pca.transform(df1)
PCA1
```

```python
PCA2= pd.DataFrame(PCA1)
PCA2.index = df1.index
PCA2.columns = ['PC1','PC2']
PCA2.to_excel('C:\\Noble\\Training\\DS Temporary Files\\Wheat_PCA.xlsx')
PCA2.head(50)
```

```python
pca.explained_variance_ratio_
```

```python
ax = PCA2.plot (kind = 'scatter' ,x ='PC2',y ='PC1', figsize=(20,10) )

for i, country in enumerate(df1.index):
    ax.annotate(country, (PCA2.iloc[i].PC2, PCA2.iloc[i].PC1))
```

```python
from sklearn.preprocessing import normalize

PCA2['country_mean'] = pd.Series(df1.mean(axis=1), index=PCA2.index)
country_mean_max = PCA2['country_mean'].max()
country_mean_min = PCA2['country_mean'].min()
country_mean_scaled = (PCA2.country_mean-country_mean_min) / country_mean_max
PCA2['country_mean_scaled'] = pd.Series(
    country_mean_scaled, 
    index=PCA2.index)
PCA2.head(20)
```

```python
PCA2.plot(kind='scatter', x='PC2', y='PC1', s=PCA2['country_mean_scaled']*100, figsize=(20,10))
```

```python
PCA2['country_sum'] = pd.Series(df1.sum(axis=1), index=PCA2.index)
country_sum_max = PCA2['country_sum'].max()
country_sum_min = PCA2['country_sum'].min()
country_sum_scaled = (PCA2.country_sum-country_sum_min) / country_sum_max
PCA2['country_sum_scaled'] = pd.Series(
    country_sum_scaled, 
    index=PCA2.index)
PCA2.plot(kind='scatter', x='PC2', y='PC1', s=PCA2['country_sum_scaled']*100, figsize=(20,10))
```

```python
PCA2['country_change'] = pd.Series(df1['2007']-df1['1990'], index=PCA2.index)
country_change_max =PCA2['country_change'].max()
country_change_min = PCA2['country_change'].min()
country_change_scaled = (PCA2.country_change - country_change_min) / country_change_max
PCA2['country_change_scaled'] = pd.Series(
    country_change_scaled, 
    index=PCA2.index)
PCA2[['country_change','country_change_scaled']].head(20)
```

```python
PCA2.plot(kind='scatter', x='PC2', y='PC1', s=PCA2['country_change_scaled']*100, figsize=(20,10))
```

```python
from sklearn.cluster import KMeans 
wcss = []
for i in range(1,8):
    kmeans = KMeans (n_clusters =  i , init = 'k-means++')
    kmeans.fit(df1)
    wcss.append (kmeans.inertia_)
wcss
```

```python
plt.plot (range(1,8),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel ('wcss')
```

```python
kmeans =  KMeans (n_clusters =  5 , init = 'k-means++')
kmeans = kmeans.fit(df1)
y_kmeans = kmeans.predict(df1)
y_kmeans
```

```python
x = PCA2.values 
plt.figure(figsize=(20,10))
plt.scatter (x[y_kmeans == 0,1],x[y_kmeans == 0,0] ,s = 100, c='r', label = 'Cluster 1')
plt.scatter (x[y_kmeans == 1,1],x[y_kmeans == 1,0] ,s = 100, c='b', label = 'Cluster 2')
plt.scatter (x[y_kmeans == 2,1],x[y_kmeans == 2,0] ,s = 100, c='g', label = 'Cluster 3')
plt.scatter (x[y_kmeans == 3,1],x[y_kmeans == 3,0] ,s = 100, c='c', label = 'Cluster 4')
plt.scatter (x[y_kmeans == 4,1],x[y_kmeans == 4,0] ,s = 100, c='m', label = 'Cluster 5')
plt.legend()
```

```python
df_y_kmeans = pd.DataFrame(y_kmeans,index=PCA2.index, columns = ['Cluster'])
df_y_kmeans
```

```python
kmeans_final = pd.concat ([df,df_y_kmeans],axis =1)
kmeans_final.to_excel('C:\\Noble\\Training\\DS Temporary Files\\kmeans_final.xlsx')
kmeans_final
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='average')
y_hc = hc.fit_predict(df1)
y_hc
```

```python
x = PCA2.values 
plt.scatter(x[y_hc==0,1], x[y_hc==0,0], s=100, c='red', label='Cluster1')
plt.scatter(x[y_hc==1,1], x[y_hc==1,0], s=100, c='blue', label='Cluster2')
plt.scatter(x[y_hc==2,1], x[y_hc==2,0], s=100, c='green', label='Cluster3')
plt.scatter(x[y_hc==3,1], x[y_hc==3,0], s=100, c='cyan', label='Cluster4')
plt.scatter(x[y_hc==4,1], x[y_hc==4,0], s=100, c='magenta', label='Cluster5')
plt.scatter(x[y_hc==5,1], x[y_hc==5,0], s=100, c='black', label='Cluster6')
plt.legend()
```
